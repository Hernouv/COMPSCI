#Defining new functions

A function definition specifies the name of a new function and the sequence of statements that run when the function is called. Here's an example:

```python
def inches_to_feet(inches):
  return inches/12
```

Now that we've defined a function, we can call it the same way we call built-in functions.

```python
p = inches_to_feet(65)
print(p)
```

#Stack call

When a function is called, Python uses call stack to keep track of where it is in the program.

```python
def func_c():
    print("Inside C")

def func_b():
    print("Inside B")
    func_c()
    print("Back in B")

def func_a():
    print("Inside A")
    func_b()
    print("Back in A")

func_a()
```

1. start func_a-->  stack: func_a
2. inside func_a() calls func_b() --> stack: func_b -> func_a
3. inside func_b() calls func_c() --> stack: func_c -> func_b -> func_a
4. func_c() finishes, back to func_b() --> stack: func_b -> func_a
5. func_b() finishes, back to func_a() --> stack: func_a
6. func_a() finish --> stack: (empty)

#Local vs. global variables

```python
counter = 0

def show_counter():
    local_counter = 1
    print("Current counter:", counter)


def increment():
    global counter
    counter = counter + 1
    print("Good_increment:", counter)

increment()
show_counter()
```

#Program testing

```python
def add(a, b):
    """
    Returns the sum of two numbers.

    Parameters:
        a (int or float): The first number.
        b (int or float): The second number.

    Returns:
        int or float: The sum of a and b.
    """
    return a - b
```

```python
def test_integers():
    assert add(2, 3) == 5, "Expected 2 + 3 to be 5"
    assert add(-1, 1) == 0, "Expected -1 + 1 to be 0"

def test_floats():
    assert add(2.5, 3.5) == 6.0, "Expected 2.5 + 3.5 to be 6.0"

def test_mixed():
    assert add(2, 3.5) == 5.5, "Expected 2 + 3.5 to be 5.5"

if __name__ == "__main__":
    test_integers()
    test_floats()
    test_mixed()
    print("All tests passed!")
```

Use assert to check the expected output of a function against the actual result.

If the condition is True, the program continues normally (test passes).

If the condition is False, Python raises an AssertionError with an optional message, indicating which test failed.

In this example, the add function contains an error: instead of adding two numbers, it subtracts them, causing an AssertionError. Change the - operator in the return statement to + to make all the tests pass.
