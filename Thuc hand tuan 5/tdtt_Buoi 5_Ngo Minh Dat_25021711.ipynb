```python
#w5a1
def Max_Of_Two(a, b):
    if a > b:
        return a
    else:
        return b
    
n1, n2 = map(int, input().split())
print(Max_Of_Two(n1, n2))
```

```python
#w5a2
def swap(a, b):
    return tuple((b, a))

n1, n2 = map(int, input().split())
swapped = swap(n1, n2)
print(swapped[0], swapped[1])
```

```python
#w5a4
def perfect_number(n):
    sum_divisors = 1
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            sum_divisors += i
            if i != n // i:
                sum_divisors += n // i
    return sum_divisors == n

n = int(input())
print(perfect_number(n))
```

```python
#w5a5
def find_pos(numbers, k):
    for i in range(len(numbers)):
        if numbers[i] == k:
            return i + 1
    return -1
        
    return -1
numbers = list(map(int, input().split()))
k = int(input())
print(find_pos(numbers, k))
```

```python
#w5a6
def fact_func(n):
    if n == 0 or n == 1:
        ret =  1
    else:
        return n * fact_func(n - 1)

n = int(input())
print(fact_func(n))
```

```python
#w5a7
def calc(n1, opt, n2):
    ret = 0
    if opt == '+':
        ret = n1 + n2
    elif opt == '-':
        ret = n1 - n2
    elif opt == '*':
        ret = n1 * n2
    elif opt == '/':
        if n2 != 0:
            ret = n1 / n2
            
    return round(ret, 2)
    
n1, opt, n2 = input().split()
n1 = float(n1)
n2 = float(n2)
print(calc(n1, opt, n2))
```

```python
#w5a8
def hamming(a, b):
    xor = a ^ b
    cnt = 0
    while xor:
        cnt += xor & 1
        xor >>= 1
        
    return cnt

a, b = map(int, input().split())
print(hamming(a, b))
```

```python
#w5a9
def digit_sum(n):
    ret = 0
    while n:
        ret += n % 10
        n //= 10
    return ret

n = int(input())
print(digit_sum(n))
```

```python
#w5a10
def check(s1, s2):
    ass = list(26 * ['0'])
    for i in range(len(s1)):
        if (ass[ord(s1[i]) - 97] == '0'):
            ass[ord(s1[i]) - 97] = s2[i]
        else:
            if (s2[i] != ass[ord(s1[i]) - 97]):
                return "false"
    
    for i in range(26):
        for j in range(i + 1, 26):
            if (ass[i] == ass[j] and ass[i] != '0'):
                return "false"
        
    return "true"

s1, s2 = map(str, input().split())
print(check(s1, s2))
```
