```python
#w10a3
def is_happy_number(n: int) -> bool:
    seen = set()
    
    while n != 1 and n not in seen:
        seen.add(n)
        n = sum(int(digit) ** 2 for digit in str(n))
    
    return n == 1


try:
    num = int(input())
    
    if is_happy_number(num):
        print("YES")
    else:
        print("NO")
        
except ValueError:
    print("Non integer")
```

```python
#w10a5
def convert_to_int_list(input_list):
    result_list = []
    
    for item in input_list:
        try:
            converted_item = int(item)
            result_list.append(converted_item)
        except (ValueError, TypeError) as e:
            print("Inconvertible")
    
    return result_list


sample_list = list(input().split())
print(sample_list)
converted_list = convert_to_int_list(sample_list)
print(converted_list)
```

```python
#w10a8
def check_file_format():
    file_name = input().strip()
    file_name = file_name.lower()
    
    # Kiểm tra định dạng file
    if file_name.endswith('.txt') or file_name.endswith('.zip'):
        print("Doc file thanh cong")
    else:
        print("File khong hop le")

check_file_format()
```

```python
#w10a9
def convert_to_integer():
    user_input = input().strip()
    
    try:
        # Thử chuyển đổi chuỗi thành số nguyên
        number = int(user_input)
        print(number)
    except ValueError:
        print("Chuoi khong hop le")

convert_to_integer()
```

```python
#w10a10
try:
    data = input().strip()
    numbers = list(map(int, data.split()))
    
    if not numbers:
        print("Mang khong hop le")
        exit(0)
    
    N = numbers[0]
    
    if N <= 0:
        print("Mang khong hop le")
        exit(0)
    
    if len(numbers) - 1 != N:
        print("Mang khong hop le")
        exit(0)
    
    arr = numbers[1:]
    
    if any(num <= 0 for num in arr):
        print("Mang khong hop le")
        exit(0)
        
    if len(set(arr)) != len(arr):
        print("Mang khong hop le")
        exit(0)
    
    total = sum(arr)
    print(total)
    
except ValueError:
    print("Mang khong hop le")
```

```python
#w10a11
import re
def is_valid_regex(pattern):
    try:
        re.compile(pattern)
        return True
    except re.error:
        return False
    
n = int(input())
for n in range(n):
    r = input()
    print(is_valid_regex(r))
    
```
