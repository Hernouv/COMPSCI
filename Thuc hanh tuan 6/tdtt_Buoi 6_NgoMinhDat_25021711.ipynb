```python
#w6a1
lst = list(map(int, input().split()))
check = {}
sz = len(lst)
id = 0
while id < sz:
    if lst[id] in check:
        del lst[id]
        sz -= 1
    else:
        check[lst[id]] = True
        id += 1
        
print(lst)
```

```python
#w6a9
dict1 = input()
dict1 = dict1[1:-1]
l1 = dict1.split(', ')
dict1 = {}
# print(l1)
dict2 = input()
dict2 = dict2[1:-1]
l2 = dict2.split(', ')
dict2 = {}

key_list = []

for rq in l1:
    p = rq.find(':')
    key = rq[:p]
    value = int(rq[p+2:])
    key_list.append(key)
    dict1[key] = value

for rq in l2:
    p = rq.find(':')
    key = rq[:p]
    value = int(rq[p+2:])
    key_list.append(key)
    dict2[key] = value
    
key_list.sort()

new_dict = {}
    
for k in key_list:
    if (k in dict1 and k in dict2):
        new_dict[k] = dict1[k] + dict2[k]
    elif k in dict1:
        new_dict[k] = dict1[k]
    else:
        new_dict[k] = dict2[k]
print(new_dict)
```

```python
#w6a10
lst = list(map(int, input().split()))
lst.sort()
k = int(input())
dict = {}
for n in lst:
    if n in dict:
        dict[n] += 1
    else:
        dict[n] = 1

res_list = []
for key in dict:
    if (int(key) > k / 2):
        break
    c = min(dict[key], dict[int(k - key)])
    if (key == k - key):
        c = dict[key] // 2
    for i in range(c):
        res_list.append((int(key), int(k - key)))
        
print(res_list)
```
