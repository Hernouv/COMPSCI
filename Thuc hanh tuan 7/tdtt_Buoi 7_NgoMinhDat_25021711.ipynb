```python
#w7a1
def bs(arr, target)->int:
    l, r = 0, len(arr)
    mid = 0
    while(l <= r):
        mid = (l + r) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            l = mid + 1
        else:
            r = mid - 1
            
    return -1
        
    
arr = list(map(int, input().split()))
v = input()
print(bs(arr, v))
```

```python
#w7a2
def cnt_occ(arr, x:int)->int:
    c = 0
    for n in arr:
        if n == x:
            c += 1
            
    return c

arr = list(map(int, input().split()))
v = int(input())
print(cnt_occ(arr, v))
```

```python
#w7a3 (bubble sort)
def bub_sort(arr):
    swap = 0
    for lim in range(len(arr), 1, -1):
        swap = 0
        for i in range(1, lim):
            if (arr[i - 1] > arr[i]):
                arr[i - 1], arr[i] = arr[i], arr[i - 1]
                swap = 1
        if (swap == 0):
            break
    return arr
    
arr = list(map(int, input().split()))
arr = bub_sort(arr)
for n in arr:
    print(n, end = ' ')
```

```python
#w7a3 (selection sort)
def sel_sort(arr):
    swap = 0
    for lim in range(len(arr), 1, -1):
        swap = 0
        pos = 0
        tv = arr[0]
        for i in range(1, lim):
            if (arr[i] >= tv):
                tv = arr[i]
                pos = i
                swap += 1
        if (swap == lim - 1):
            break
        arr[pos], arr[lim - 1] = arr[lim - 1], arr[pos]
    return arr
    
arr = list(map(int, input().split()))
arr = sel_sort(arr)
for n in arr:
    print(n, end = ' ')
```

```python
#w7a4
def fin(arr)->tuple:
    d = {}
    for i in range(len(arr)):
        if arr[i] in d:
            d[arr[i]][1] += 1
        else:
            d[arr[i]] = [i, 1]
    
    r = (0, 0, 0) #value, start_pos, freq
    
    for val, info in d.items():
        if (info[1] > r[2]):
            r = (val, info[0], info[1])
        elif (info[1] == r[2] and info[0] < r[1]):
            r = (val, info[0], info[1])
    
    return (r[0], r[2])
            

arr = list(map(int, input().split()))
res = fin(arr) 
print(f"{res[0]} xuat hien nhieu nhat, som nhat, {res[1]} lan")
```

```python
#w7a5
arr = []
def quicksort(arr, l, r):
    if (l >= r):
        return
    p = arr[(l + r) // 2]
    i, j = l, r
    while (i < j):
        while (arr[i] < p):
            i += 1
        while (arr[j] > p):
            j -= 1
        if (i <= j):
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
            j -= 1
            
    quicksort(arr, i, r)
    quicksort(arr, l, j)
    
arr = list(map(int, input().split()))
s = int(input())
quicksort(arr, 0, len(arr) - 1)
# print(arr)
L, R = 0, len(arr) - 1
sz = R

res = 0

while (L < R):
    while (arr[L] + arr[R] != s):
        if (arr[L] + arr[R] < s):
            while(arr[L + 1] == arr[L] and L < sz):
                L += 1
            L += 1
        if (arr[L] + arr[R] > s):
            while (arr[R - 1] == arr[L] and R > 0):
                R -= 1
            R -= 1
    
    if (arr[L] + arr[R] == s):
        # print(L, R)
        v1, v2 = arr[L], arr[R]
        c1, c2 = 1, 1
        while (arr[L + 1] == arr[L] and L < sz):
            L += 1
            c1 += 1
        while (arr[R - 1] == arr[R] and R > 0):
            R -= 1
            c2 += 1
        
        # print(f"c: {c1} {c2}")
            
        if (v1 == v2):
            res += c1 * (c2 + 1) // 2
        else:
            res += c1 * c2
        
        L += 1
        R -= 1

print(res)
```

```python
#w7a8
i_in = input().split("], [")
i_in[0] = i_in[0][2:]
i_in[-1] = i_in[-1][:-3]
q_in = input().split(", ")
q_in[0] = q_in[0][1:]
q_in[-1] = q_in[-1][:-1]

intervals = []
queries = []

for s in i_in:
    tn = 0
    ta = []
    for c in s:
        if '0' <= c <= '9':
            tn = tn * 10 + int(c)
        elif c == ',':
            ta.append(tn)
            tn = 0
    ta.append(tn)
    intervals.append([ta[0], ta[1]])

for s in q_in:
    queries.append(int(s))
    
intervals.sort()

res = []
for q in queries:
    ans = 1e18
    for d in intervals:
        if (d[0] <= q):
            if (d[1] >= q):
                ans = min(ans, d[1] - d[0] + 1)
        else:
            break
    if (ans == 1e18):
        res.append(-1)
    else:
        res.append(ans)

print(res)                
```
