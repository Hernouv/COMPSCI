```python
#w8a1
class cylinder:
    def __init__(self, r, h):
        self.r = float(r)
        self.h = float(h)
    
    def volume(self):
        return round(3.14 * self.r ** 2 * self.h, 2)

    def area(self):
        return round(2 * self.r * 3.14 * self.h + 3.14 * self.r ** 2 * 2, 2)

f = cylinder(*map(int, input().split()))
print(f.volume())
print(f.area())
```

```python
#w8a2
import math
class calendar:
    def __init__(self, day, month, year):
        self.day = int(day)
        self.month = int(month)
        self.year = int(year)
    
    maxdate = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    
    def valid_check(self):
        t = 0
        if (self.month == 2):
            t = t = int(not bool(self.year % 4))
        if (self.day <= calendar.maxdate[self.month] + t):
            return True
        return False
    
    def tmr(self):
        if (not calendar.valid_check(self)):
            return "INVALID"
        
        t = 0
        if (self.month == 2):
            t = int(not bool(self.year % 4))
        if (self.day == calendar.maxdate[self.month] + t):
            if (self.month == 12):
                self.day, self.month = 1, 1
                self.year += 1
            else:
                self.day = 1
                self.month += 1
        else:
            self.day += 1
        
        if (self.day < 10):
            self.day = f"0{self.day}"
        if (self.month < 10):
            self.month = f"0{self.month}"
        if (self.year < 1000):
            self.year = f"{self.year}"
            self.year = '0' * (4 - len(self.year)) + self.year
            
        return f"{self.day}/{self.month}/{self.year}"
            
date = input().split('/')
print(calendar.tmr(calendar(*map(int, date))))
        
```

```python
#w8a5
class ShoppingCart:
    def __init__(self):
        self.cart = []

    def add_item(self, name, price):
        item = {
            "name": name,
            "price": price
        }
        self.cart.append(item)
        print(f"--> Đã thêm '{name}' với giá {price} vào giỏ hàng.")

    def remove_item(self, name):
        found = False
        for item in self.cart[:]: 
            if item['name'] == name:
                self.cart.remove(item)
                found = True
                print(f"--> Đã xóa sản phẩm '{name}' khỏi giỏ hàng.")
                break
        
        if not found:
            print(f"--> Không tìm thấy sản phẩm tên '{name}' trong giỏ.")

    def is_empty(self):
        return len(self.cart) == 0

    def calculate_total(self):
        total = 0
        for item in self.cart:
            total += item['price']
        return total

    def display_cart(self):
        if self.is_empty():
            print("--> Giỏ hàng đang trống.")
        else:
            print("\n--- DANH SÁCH SẢN PHẨM TRONG GIỎ ---")
            for index, item in enumerate(self.cart, 1):
                print(f"{index}. Tên: {item['name']} - Giá: {item['price']}")
            print("------------------------------------")

    def clear_cart(self):
        self.cart = []
        print("--> Đã xóa toàn bộ giỏ hàng.")

def main():
    my_cart = ShoppingCart() 

    while True:
        print("\n" + "="*30)
        print("   QUẢN LÝ GIỎ HÀNG (MENU)")
        print("="*30)
        print("2. Thêm sản phẩm vào giỏ")
        print("3. Xóa sản phẩm theo tên")
        print("4. Kiểm tra giỏ hàng rỗng")
        print("5. Tính tổng tiền")
        print("6. Hiển thị giỏ hàng")
        print("7. Xóa toàn bộ giỏ hàng")
        print("="*30)

        choice = input("Nhập lựa chọn của bạn (2-7): ")

        if choice == '2':
            name = input("Nhập tên sản phẩm: ")
            try:
                price = float(input("Nhập giá sản phẩm: "))
                my_cart.add_item(name, price)
            except ValueError:
                print("--> Lỗi: Giá phải là một con số.")
        
        elif choice == '3':
            name = input("Nhập tên sản phẩm cần xóa: ")
            my_cart.remove_item(name)
        
        elif choice == '4':
            if my_cart.is_empty():
                print("--> Giỏ hàng ĐANG RỖNG.")
            else:
                print("--> Giỏ hàng KHÔNG RỖNG.")
        
        elif choice == '5':
            total = my_cart.calculate_total()
            print(f"--> Tổng tiền giỏ hàng là: {total}")
        
        elif choice == '6':
            my_cart.display_cart()
        
        elif choice == '7':
            confirm = input("Bạn có chắc chắn muốn xóa hết không? (y/n): ")
            if confirm.lower() == 'y':
                my_cart.clear_cart()
        
        else:
            print("--> Lựa chọn không hợp lệ. Vui lòng chọn từ 2 đến 7.")

        exit_choice = input("\nCó muốn thoát khỏi chương trình không? (nhập 'yes' để thoát): ")
        if exit_choice.lower() == 'yes':
            print("Cảm ơn bạn đã sử dụng chương trình. Tạm biệt!")
            break

if __name__ == "__main__":
    main()
```

```python
#w8a6
class Rectangle:
    def __init__(self, w, h, k):
        self.w = w
        self.h = h
        self.k = k
        
    def area(self):
        return round(self.w * self.h, 2)
    
    def para(self):
        return round(2 * (self.w + self.h), 2)
    
    def scale(self):
        self.w *= self.k
        self.h *= self.k
        
rect = Rectangle(*map(float, input().split()))
rect.scale()
print(f"{rect.area()} {rect.para()}")
```

```python
#w8a7
class simplify():
    def __init__(self, num, den):
        self.num = num
        self.den = den
        
    def gcd(self)->int:
        n1, n2 = abs(self.num), abs(self.den)
        while (n1 and n2):
            if (n1 > n2):
                n1 %= n2
            else:
                n2 %= n1
        return max(n1, n2)
    
    def ret(self):
        g = self.gcd()
        return self.num // g, self.den // g

class Fraction():
    def __init__(self, num, den):
        self.num, self.den = simplify.ret(simplify(num, den))
        
    def __str__(self):
        return f"{self.num}/{self.den}"
    
def calc(n1, op, n2):
    if (op == '+'):
        rn = n1.num * n2.den + n2.num * n1.den
        rd = n1.den * n2.den
        rn, rd = simplify.ret(simplify(rn, rd))
        return f"{rn}/{rd}"
    elif (op == '-'):
        rn = n1.num * n2.den - n2.num * n1.den
        rd = n1.den * n2.den
        rn, rd = simplify.ret(simplify(rn, rd))
        return f"{rn}/{rd}"
    elif (op == '*'):
        rn = n1.num * n2.num
        rd = n1.den * n2.den
        rn, rd = simplify.ret(simplify(rn, rd))
        return f"{rn}/{rd}"
    elif (op == '/'):            
        rn = n1.num * n2.den
        rd = n1.den * n2.num
        rn, rd = simplify.ret(simplify(rn, rd))
        return f"{rn}/{rd}"
            
a, b, op, c, d = map(str, input().split())
n1 = Fraction(int(a), int(b))
n2 = Fraction(int(c), int(d))
print(calc(n1, op, n2))

```

```python
#w8a8
class BankAccount():
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)
    
    def __str__(self):
        return f"{self.owner}: {self.balance:2d}"
    
    def proc(self, rq, am):
        if (rq == "DEPOSIT"):
            self.balance += am
        elif (rq == "WITHDRAW"):
            if (self.balance >= am):
                self.balance = am
    
    
acc = BankAccount(*map(str, input().split()))
n = int(input())
for i in range(n):
    rq, am = map(str, input().split())
    am = float(am)
    acc.proc(rq, am)

print(acc.balance)
```

```python
#w8a10
class Book():
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year 
    
    def __str__(self):
        return f"{self.title} - {self.author} - {self.year}"
    
class Library():
    def __init__(self):
        self.a_lib = {}
        self.y_lib = {}
    
    def add(self, info):
        try:
            self.a_lib[info.author].append((info.title, info.year))
        except:
            self.a_lib[info.author] = [(info.title, info.year)]
            
        try:
            self.y_lib[info.year].append((info.title, info.author))
        except:
            self.y_lib[info.year] = [(info.title, info.author)]
    
    def count_by_author(self, name):
        try:    
            return len(self.a_lib[name])
        except:
            return 0
    def find_by_year(self, y):
        try:
            return len(self.y_lib[y])
        except:
            return 0
    
        
        
n = int(input())
lib = Library()
for i in range(n):
    rq, cont = map(str, input().split())
    cont = cont.split(';')
    if (rq == 'ADD'):
        book = Book(cont[0], cont[1], cont[2])
        lib.add(book)
    elif (rq == "COUNT"):
        print(lib.count_by_author(cont[0]))
    elif (rq == "COUNTYEAR"):
        print(lib.find_by_year(cont[0]))
```

```python
#w8a11
import math

class Points():
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __sub__(self, no):
        return Points(self.x - no.x, self.y - no.y, self.z - no.z)

    def dot(self, no):
        return self.x * no.x + self.y * no.y + self.z * no.z

    def cross(self, no):
        return Points(
            self.y * no.z - self.z * no.y,
            self.z * no.x - self.x * no.z,
            self.x * no.y - self.y * no.x
        )

    def absolute(self):
        return pow((self.x ** 2 + self.y ** 2 + self.z ** 2), 0.5)


points = list()
for i in range(4):
    a = list(map(float, input().split()))
    points.append(Points(a[0], a[1], a[2]))

a, b, c, d = points[0], points[1], points[2], points[3]

ab = b - a
bc = c - b
cd = d - c

x = ab.cross(bc) 
y = bc.cross(cd)

cos_phi = x.dot(y) / (x.absolute() * y.absolute())

cos_phi = max(min(cos_phi, 1.0), -1.0)

res = math.degrees(math.acos(cos_phi))

print(round(res, 2))
```

```python
#w8a12
class CNum:
    def __init__(self, initRealPart, initImaginaryPart):
        self.re = initRealPart
        self.imt = initImaginaryPart

    def multiply(self, other):
        res_re = (self.re * other.re) - (self.imt * other.imt)
        res_im = (self.re * other.imt) + (self.imt * other.re)
        return CNum(res_re, res_im)

try:
    line1 = input().split()
    c1 = CNum(int(line1[0]), int(line1[1]))

    line2 = input().split()
    c2 = CNum(int(line2[0]), int(line2[1]))

    result = c1.multiply(c2)

    print(f"{result.re} {result.imt}")

except EOFError:
    pass
```

```python
#w8a13
class Triangle:
    def __init__(self, e1, e2, e3):
        self.e1 = e1
        self.e2 = e2
        self.e3 = e3
    
    def getArea(self):
        half_p = (self.e1 + self.e2 + self.e3) / 2
        return round((half_p * (half_p - self.e1)* (half_p - self.e2)* (half_p - self.e3)) ** 0.5, 2)
        
def check_valid(tri_inp):
    arr = [tri_inp.e1, tri_inp.e2, tri_inp.e3]
    if (arr[0] < 0 or arr[1] < 0 or arr[2] < 0):
        return False
    
    if (arr[0] + arr[1] <= arr[2]):
        return False
    elif (arr[0] + arr[2] <= arr[1]):
        return False
    elif (arr[1] + arr[2] <= arr[0]):
        return False
    
    return True

tri_inp = Triangle(*map(int, input().split()))

if (check_valid(tri_inp)):
    print(tri_inp.getArea())
else:
    print("invalid")
```
