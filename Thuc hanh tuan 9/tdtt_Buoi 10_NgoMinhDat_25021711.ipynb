```python
#w9a1
def merge(a1, a2):
    ret = []
    i, j = 0, 0
    while (i < len(a1) and j < len(a2)):
        if (a1[i] < a2[j]):
            ret.append(a1[i])
            i += 1
        else:
            ret.append(a2[j])
            j += 1
            
    ret.extend(a1[i:])
    ret.extend(a2[j:])
    return ret
    
arr1 = list(map(int, input().split()))
arr2 = list(map(int, input().split()))
res = merge(arr1, arr2)
for n in res:
    print(n, end = ' ')
```

```python
#w9a2
def merge(a1, a2):
    ret = []
    i, j = 0, 0
    while (i < len(a1) and j < len(a2)):
        if (a1[i] < a2[j]):
            ret.append(a1[i])
            i += 1
        else:
            ret.append(a2[j])
            j += 1
            
    ret.extend(a1[i:])
    ret.extend(a2[j:])        
    return ret

matrx = []
pos = 0
while True:
    try:
        row_inp = input().split()
    except:
        break
    
    pos = int(row_inp[-1])
    if row_inp:
        row = list(map(int, row_inp))
        if (not matrx):
            matrx.extend(row)
        else:
            matrx = merge(matrx, row)
    else:
        break

if (matrx[pos - 1] == pos): pos += 1
print(matrx[pos - 1])
```

```python
#w9a3
p1, p2 = -1, -1
def dswap(arr, idx):
    if (idx == len(arr) - 1):
        return [-1]
    
    tv, tp = 10, -1
    for i in range(idx + 1, len(arr)):
        if (tv > arr[i]):
            tv = arr[i]
            tp = i
        
    if (tv < arr[idx]):
        arr[idx], arr[tp] = arr[tp], arr[idx]
        return arr
    else:
        if (tv == arr[idx]):
            p1, p2 = tp, idx    
        return dswap(arr, idx + 1)

n = int(input())
a = []
while(n):
    a.append(n % 10)
    n //= 10
a = a[::-1]
# print(a)
res = dswap(a, 0)

if (res[0] == -1):
    if (p1 != -1 and p2 != -1):
        a[p1], a[p2] = a[p2], a[p1]
    else:
        a[-1], a[-2] = a[-2], a[-1]

for d in res:
    n *= 10
    n += d
print(n)
```

```python
#w9a4
def scan(s1, s2):
    i, j = 0, 0
    while (i < len(s1) and j < len(s2)):
        if (s1[i] == s2[j] or ord(s1[i]) == ord(s2[j]) + 32):
            j += 1
        i += 1
    if (j == len(s2)):
        return "YES"
    else:
        return "NO"
        

n = int(input())
for i in range(n):
    s1 = input()
    s2 = input()
    print(scan(s1, s2))
```

```python
#w9a6
def find_i(a, index):
    if index < 0:
        return -1
    if a[index] < a[index + 1]:
        return index
    return find_i(a, index - 1)

def find_j(a, i, index):
    if a[index] > a[i]:
        return index
    return find_j(a, i, index - 1)

def reverse_tail(a, start, end):
    if start >= end:
        return
    a[start], a[end] = a[end], a[start]
    reverse_tail(a, start + 1, end - 1)

def next_perm():
    n = int(input())
    a = list(map(int, input().split()))

    i = find_i(a, n - 2)

    if i == -1:
        a.sort() 
    else:
        j = find_j(a, i, n - 1)
        a[i], a[j] = a[j], a[i]
        reverse_tail(a, i + 1, n - 1)
        
    print(*(a))
    
next_perm()
```

```python
#w9a7
def canPlaceFlowers(flowerbed, k):
    res = 0
    t = [-2]
    for i in range(len(flowerbed)):
        if (flowerbed[i] == 1):
            t.append(i)
    t.append(len(flowerbed) + 1)

    for i in range(1, len(t)):
        v = t[i] - t[i - 1] - 3
        if (v > 0):
            res += v
    
    return (res >= k)
    
a = list(map(int, input().split()))
n = int(input())
print(canPlaceFlowers(a, n))
```

```python
#w9a8

visited = set()
l = 0
def is_path_crossing(moves, nx = 0, ny = 0, id = 0):
    if (id == l):
        return False
    
    visited.add((nx, ny))
    if (moves[id] == 'R'):
        nx += 1
    elif (moves[id] == 'L'):
        nx -= 1
    elif (moves[id] == 'U'):
        ny += 1
    elif (moves[id] == 'D'):
        ny -= 1
    
    if ((nx, ny) in visited):
        return True
    return is_path_crossing(moves, nx, ny, id + 1)
    
    
m = input()
l = len(m)
print(is_path_crossing(m))
```

```python
#w9a9
def median(nums):
    a, b, c, d, e = nums
    if a > b:
        a, b = b, a
    if c > d:
        c, d = d, c
    
    if a > c:
        a, c = c, a
        b, d = d, b
    
    if b > e:
        b, e = e, b
    if b > c:
        b, c = c, b
        d, e = e, d
    if c > e:
        return d if d < e else e
    else:
        return c if c < d else d

a = list(map(int, input().split()))
print (median(a))
        
```
